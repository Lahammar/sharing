```python
plt.bar(manader, k2020["Kuukauden sadesumma (mm)"], width=0.4)
plt.bar(manader[:8], k2021["Kuukauden sadesumma (mm)"], align="edge",width=0.4)
```

```python
fig = plt.figure(figsize = (15,5))

plt.plot(manader, k2020["Kuukauden keskilämpötila (degC)"], label = 2020)
plt.bar(manader, k2020["Kuukauden sadesumma (mm)"]*0.3, width=0.4, alpha=0.3, bottom=-10)

plt.plot(manader[0:8], k2021["Kuukauden keskilämpötila (degC)"], label = 2021)
plt.bar(manader[:8], k2021["Kuukauden sadesumma (mm)"]*0.3 ,width=0.4,alpha=0.3, align="edge", bottom=-10)


plt.title("Månadernas medeltemperaturer och jämförelse av regnmängder", fontsize=16)
plt.ylabel("temperatur, $^\circ C$ ",fontsize=12)
plt.legend(fontsize=14)

plt.show()
```
